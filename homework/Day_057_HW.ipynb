### 用 iris (dataset.load_iris()) 資料嘗試跑 hierarchical clustering

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import AgglomerativeClustering

np.random.seed(5)
```

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data[:,:-1]
y = iris.target
features_names = iris['feature_names'][:-1]
```

```python
print(iris['DESCR'])
```

```python
def plot_func(data, target, labels):
    fig = plt.figure(figsize=(8,6))
    ax = fig.gca(projection='3d')
    ax.scatter(data[:, 0], data[:, 1], data[:, 2],
               c=target, cmap=plt.cm.rainbow)
    ax.set_xlabel(labels[0])
    ax.set_ylabel(labels[1])
    ax.set_zlabel(labels[2])
    ax.view_init(elev=25, azim=45)
```

```python
def hierarchical_estimator(n_clusters, data, linkage):
    est = AgglomerativeClustering(n_clusters=n_clusters,linkage=linkage)
    est.fit(data)
    return est.labels_
```

```python
### original data
plot_func(X, y, features_names)
```

```python
# 3 clusters 
plot_func(X, hierarchical_estimator(3, X ,'ward'), features_names)

# 3 clusters
plot_func(X, hierarchical_estimator(3, X ,'complete'), features_names)

# 3 clusters
plot_func(X, hierarchical_estimator(4, X ,'average'), features_names)
```
