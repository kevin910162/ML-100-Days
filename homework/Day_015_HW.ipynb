```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix1 = np.random.uniform(-1, 1, size = (nrow,ncol))

indice1 = np.random.choice([0,1,2], size=nrow)
plot_data1 = pd.DataFrame(matrix1, indice1)


grid = sns.PairGrid(data = plot_data1, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag( plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
